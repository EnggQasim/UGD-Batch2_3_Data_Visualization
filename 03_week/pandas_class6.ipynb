# How to create pandas series?

* list
* tuple
* dictionary
* list with addtional index list


```python
# list
l1 = ['a','b','c']
l1 = pd.Series(l1)
l1
```

```python
# tuple
l1 = ('a','b','c')
l1 = pd.Series(l1)
l1
```

```python
# set
l1 = {'a','b','c'}
l1 = pd.Series(l1)
l1
```

```python
l1 = ['a','b','c']
i1 = ['x','y','z']

s1 = pd.Series(data=l1, index=i1)
s1
```

```python
s1 = pd.Series({
    "x":'a',
    "y":'abc',
    "z":'pakistan'
})

s1
```

```python
l1 =  [1,2,3]
pd.Series(l1)
```

```python
l1 =  [1,2,3]
pd.Series(l1, dtype=np.float64)
```

```python
l1 =  [1,2,3]
pd.Series(l1, dtype=np.float64, name='sid')
```

# how to create dataframe?
